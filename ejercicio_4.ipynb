Importación de librerias

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import stanza
stanza.download('es')

# Cargar el pipeline de español
nlp = stanza.Pipeline('es', processors='tokenize,mwt,pos,lemma,depparse')

```

```python
with open('./Ejercicio_4/documento_1.txt', 'r', encoding='utf-8', errors='replace') as file:
    documento_1 = file.read()

with open('./Ejercicio_4/documento_2.txt', 'r', encoding='utf-8', errors='replace') as file:
    documento_2 = file.read()  
```

Normalizacion de datos:

Eliminar etiquetas HTML, URLs, puntuación, emojis y espacios extra.

```python
def normalizar_texto(texto):
    texto = re.sub(r'<[^>]+>', '', texto)              # Eliminar etiquetas HTML
    texto = re.sub(r'http\S+', '', texto)              # Eliminar URLs
    texto = re.sub(r'[^\w\s]', '', texto)              # Eliminar puntuación
    texto = re.sub(r'\s+', ' ', texto).strip()         # Espacios extra
    return texto
```

Carga los nuevos documentos

```python
documento_1_normalizado = normalizar_texto(documento_1)
documento_2_normalizado = normalizar_texto(documento_2)
```

Mostrar documento 1 y 2 normalizados

```python
print(documento_1_normalizado)
print(documento_2_normalizado)
```

Estadísticas básicas sin Stanza 

Contar párrafos, frases y palabras.

```python
def estadisticas_basicas(texto):
    parrafos = texto.split('\n')
    oraciones = re.split(r'[.!?]', texto)
    palabras = texto.split()

    return {
        "párrafos": len([p for p in parrafos if p.strip() != '']),
        "oraciones": len([o for o in oraciones if o.strip() != '']),
        "palabras": len(palabras)
    }

print("Documento 1:", estadisticas_basicas(documento_1_normalizado))
print("Documento 2:", estadisticas_basicas(documento_2_normalizado))
```

 Tokenización, stopwords y lematización con Stanza sobre documentos normalizados

```python
from collections import Counter
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stopwords_es = set(stopwords.words('spanish'))

def procesar_stanza(texto):
    doc = nlp(texto)
    lemas = []
    for sent in doc.sentences:
        for word in sent.words:
            if word.lemma not in stopwords_es and word.upos != 'PUNCT':
                lemas.append(word.lemma.lower())
    return lemas

# Aplicar a los textos
tokens1 = procesar_stanza(documento_1_normalizado)
tokens2 = procesar_stanza(documento_2_normalizado)

# Palabras más frecuentes
print("Frecuentes texto 1:", Counter(tokens1).most_common(10))
print("Frecuentes texto 2:", Counter(tokens2).most_common(10))

# Nubes de palabras
def generar_nube(tokens, titulo):
    texto = ' '.join(tokens)
    nube = WordCloud(width=800, height=400, background_color='white').generate(texto)
    plt.figure(figsize=(10, 5))
    plt.imshow(nube, interpolation='bilinear')
    plt.axis('off')
    plt.title(titulo)
    plt.show()

generar_nube(tokens1, "Nube de Documento 1")
generar_nube(tokens2, "Nube de Documento 2")
```

Tokenización, stopwords y lematización con Stanza sobre documentos no normalizados

```python
from collections import Counter
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stopwords_es = set(stopwords.words('spanish'))

def procesar_stanza(texto):
    doc = nlp(texto)
    lemas = []
    for sent in doc.sentences:
        for word in sent.words:
            if word.lemma not in stopwords_es and word.upos != 'PUNCT':
                lemas.append(word.lemma.lower())
    return lemas

# Aplicar a los textos
tokens1 = procesar_stanza(documento_1)
tokens2 = procesar_stanza(documento_2)

# Palabras más frecuentes
print("Frecuentes texto 1:", Counter(tokens1).most_common(10))
print("Frecuentes texto 2:", Counter(tokens2).most_common(10))

# Nubes de palabras
def generar_nube(tokens, titulo):
    texto = ' '.join(tokens)
    nube = WordCloud(width=800, height=400, background_color='white').generate(texto)
    plt.figure(figsize=(10, 5))
    plt.imshow(nube, interpolation='bilinear')
    plt.axis('off')
    plt.title(titulo)
    plt.show()

generar_nube(tokens1, "Nube de Documento 1")
generar_nube(tokens2, "Nube de Documento 2")
```

Etiquetado gramatical y nubes por categoría (Sustantivos, Verbos, Adjetivos)

```python
def etiquetas_por_categoria(doc):
    categorias = {'NOUN': [], 'VERB': [], 'ADJ': []}
    for sent in doc.sentences:
        for word in sent.words:
            if word.upos in categorias:
                categorias[word.upos].append(word.lemma.lower())
    return categorias

doc1 = nlp(documento_1_normalizado)
doc2 = nlp(documento_2_normalizado)

etiquetas1 = etiquetas_por_categoria(doc1)
etiquetas2 = etiquetas_por_categoria(doc2)

# Mostrar nubes por categoría
for cat in etiquetas1:
    generar_nube(etiquetas1[cat], f"{cat} - Documento 1")
    generar_nube(etiquetas2[cat], f"{cat} - Documento 2")
```

Extracción de tripletas SVO (Sujeto – Verbo – Objeto)

```python
def extraer_svo(doc):
    tripletas = []
    for sent in doc.sentences:
        subj = ""
        verb = ""
        obj = ""
        for word in sent.words:
            if word.deprel == 'nsubj':
                subj = word.text
                verb = sent.words[word.head - 1].text if word.head > 0 else ""
            if word.deprel == 'obj' and verb:
                obj = word.text
                tripletas.append((subj, verb, obj))
    return tripletas

svo1 = extraer_svo(doc1)
svo2 = extraer_svo(doc2)

print("SVO Documento 1:", svo1[:5])
print("SVO Documento 2:", svo2[:5])
```
